## **Word by Word OCR with Confidence Scores**

We need to implement `TSV` i.e. Tab Separated Values output format to get word by word OCR with confidence scores.

So the plan is:

- Ask for `TSV` output format from Tesseract OCR.

- Parse the `TSV` output to extract words and their corresponding confidence scores.

- Keep only rows where the level is `5` (which corresponds to words).

- Extract the word text and confidence score from those rows.
